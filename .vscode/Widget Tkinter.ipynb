## Widget Tkinter 

### Daftar Widget yang ada di Tkinter
#### - Part 1

```python
# Import library tkinter dan modul tambahan
import tkinter as tk
from tkinter import ttk, messagebox

# Fungsi untuk aksi pada tombol
def on_button_click():
    messagebox.showinfo("Button Clicked", "Tombol berhasil diklik!")

# Fungsi utama untuk membuat semua widget
def create_widgets():
    # Membuat jendela utama
    root = tk.Tk()
    root.title("Contoh Semua Widget Tkinter")
    root.geometry("600x800")
    
    # 1. Penjelasan Label
    tk.Label(root, text="1. Label: Menampilkan teks statis", font=("Arial", 12)).pack(pady=5)
    tk.Label(root, text="Ini adalah Label", font=("Arial", 12)).pack(pady=5)

    # 2. Button
    tk.Label(root, text="2. Button: Tombol interaktif", font=("Arial", 12)).pack(pady=5)
    tk.Button(root, text="Klik Saya", command=on_button_click).pack(pady=5)

    # 3. Entry
    tk.Label(root, text="3. Entry: Input teks satu baris", font=("Arial", 12)).pack(pady=5)
    tk.Entry(root, width=30).pack(pady=5)

    # 4. Text
    tk.Label(root, text="4. Text: Input teks multi-baris", font=("Arial", 12)).pack(pady=5)
    text_box = tk.Text(root, height=4, width=40)
    text_box.pack(pady=5)
    text_box.insert("1.0", "Ini adalah widget Text (multi-line).")

    # 5. Frame
    tk.Label(root, text="5. Frame: Kontainer untuk widget lain", font=("Arial", 12)).pack(pady=5)
    frame = tk.Frame(root, bg="lightblue", width=200, height=50)
    frame.pack(pady=5)
    tk.Label(frame, text="Ini adalah Frame", bg="lightblue").pack()

    # 6. Checkbutton
    tk.Label(root, text="6. Checkbutton: Pilihan ya/tidak", font=("Arial", 12)).pack(pady=5)
    check_var = tk.IntVar()
    tk.Checkbutton(root, text="Centang Saya", variable=check_var).pack(pady=5)

    # 7. Radiobutton
    tk.Label(root, text="7. Radiobutton: Pilihan tunggal dari beberapa opsi", font=("Arial", 12)).pack(pady=5)
    radio_var = tk.StringVar(value="1")
    tk.Radiobutton(root, text="Pilihan 1", value="1", variable=radio_var).pack()
    tk.Radiobutton(root, text="Pilihan 2", value="2", variable=radio_var).pack()

    # 8. Listbox
    tk.Label(root, text="8. Listbox: Menampilkan daftar item", font=("Arial", 12)).pack(pady=5)
    listbox = tk.Listbox(root, height=4)
    listbox.pack(pady=5)
    for item in ["Item 1", "Item 2", "Item 3"]:
        listbox.insert(tk.END, item)

    # 9. Scale
    tk.Label(root, text="9. Scale: Slider untuk memilih nilai numerik", font=("Arial", 12)).pack(pady=5)
    tk.Scale(root, from_=0, to=100, orient="horizontal").pack(pady=5)

    # 10. Spinbox
    tk.Label(root, text="10. Spinbox: Input angka dengan kontrol naik/turun", font=("Arial", 12)).pack(pady=5)
    tk.Spinbox(root, from_=0, to=10, width=5).pack(pady=5)

    # Menjalankan aplikasi
    root.mainloop()

# Jalankan fungsi untuk membuat GUI
create_widgets()
```

```python
# Import library tkinter dan modul tambahan
import tkinter as tk
from tkinter import ttk, messagebox

# Fungsi untuk aksi pada tombol
def on_button_click():
    messagebox.showinfo("Button Clicked", "Tombol berhasil diklik!")

# Fungsi utama untuk membuat semua widget
def create_widgets():
    # Membuat jendela utama
    root = tk.Tk()
    root.title("Contoh Semua Widget Tkinter")
    root.geometry("600x800")
    
    # 11. Canvas
    tk.Label(root, text="11. Canvas: Untuk menggambar bentuk 2D", font=("Arial", 12)).pack(pady=5)
    canvas = tk.Canvas(root, width=200, height=100, bg="white")
    canvas.pack(pady=5)
    canvas.create_line(0, 0, 200, 100, fill="blue")
    canvas.create_oval(50, 20, 150, 80, fill="red")

    # 12. Menu
    tk.Label(root, text="12. Menu: Bar navigasi di bagian atas", font=("Arial", 12)).pack(pady=5)
    menu_bar = tk.Menu(root)
    file_menu = tk.Menu(menu_bar, tearoff=0)
    file_menu.add_command(label="Open")
    file_menu.add_command(label="Save")
    file_menu.add_separator()
    file_menu.add_command(label="Exit", command=root.quit)
    menu_bar.add_cascade(label="File", menu=file_menu)
    root.config(menu=menu_bar)

    # 13. Combobox
    tk.Label(root, text="13. Combobox: Dropdown list", font=("Arial", 12)).pack(pady=5)
    ttk.Combobox(root, values=["Option 1", "Option 2", "Option 3"]).pack(pady=5)

    # 14. Treeview
    tk.Label(root, text="14. Treeview: Menampilkan data dalam tabel", font=("Arial", 12)).pack(pady=5)
    tree = ttk.Treeview(root, columns=("col1", "col2"), show="headings", height=4)
    tree.pack(pady=5)
    tree.heading("col1", text="Column 1")
    tree.heading("col2", text="Column 2")
    tree.insert("", "end", values=("Data 1", "Data 2"))

    # 15. Progressbar
    tk.Label(root, text="15. Progressbar: Menampilkan status progres", font=("Arial", 12)).pack(pady=5)
    progress = ttk.Progressbar(root, orient="horizontal", length=200, mode="determinate")
    progress.pack(pady=5)
    progress["value"] = 70

    # 16. Notebook
    tk.Label(root, text="16. Notebook: Menampilkan beberapa tab", font=("Arial", 12)).pack(pady=5)
    notebook = ttk.Notebook(root)
    tab1 = ttk.Frame(notebook)
    tab2 = ttk.Frame(notebook)
    notebook.add(tab1, text="Tab 1")
    notebook.add(tab2, text="Tab 2")
    notebook.pack(pady=5)

    # 17. LabelFrame
    tk.Label(root, text="17. LabelFrame: Kontainer dengan label", font=("Arial", 12)).pack(pady=5)
    labelframe = tk.LabelFrame(root, text="LabelFrame Contoh", padx=10, pady=10)
    labelframe.pack(pady=5)
    tk.Label(labelframe, text="Ini adalah LabelFrame").pack()

    # 18. PanedWindow
    tk.Label(root, text="18. PanedWindow: Membagi jendela menjadi beberapa panel", font=("Arial", 12)).pack(pady=5)
    panedwindow = tk.PanedWindow(root)
    panedwindow.pack(pady=5)
    left = tk.Label(panedwindow, text="Left Pane", bg="lightgray")
    right = tk.Label(panedwindow, text="Right Pane", bg="lightgray")
    panedwindow.add(left)
    panedwindow.add(right)

    # 19. Separator
    tk.Label(root, text="19. Separator: Garis horizontal atau vertikal", font=("Arial", 12)).pack(pady=5)
    ttk.Separator(root, orient="horizontal").pack(fill="x", pady=5)

    # Menjalankan aplikasi
    root.mainloop()

# Jalankan fungsi untuk membuat GUI
create_widgets()
```

```python
import tkinter as tk
from tkinter import StringVar

# 20. Option Menu
# Create the main window
root = tk.Tk()
root.title("OptionMenu Example")

# Define the options
options = ["Option 1", "Option 2", "Option 3", "Option 4"]

# Create a StringVar to hold the selected option
selected_option = StringVar()
selected_option.set(options[0])  # Set the default option

# Create the OptionMenu
option_menu = tk.OptionMenu(root, selected_option, *options)
option_menu.pack(pady=20)

# Function to display the selected option
def show_selection():
    print(f"Selected option: {selected_option.get()}")

# Button to print the selected option
btn = tk.Button(root, text="Show Selection", command=show_selection)
btn.pack(pady=10)

# Run the application
root.mainloop()
```
